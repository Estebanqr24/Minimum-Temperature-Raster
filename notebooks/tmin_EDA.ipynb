```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer el archivo CSV que se creó al procesar la data
df = pd.read_csv('../data/processed/tmin_zonal_distritos.csv')

# Vista rápida de estadísticos de la media
df.head()
```

```python
# Estadísticos descriptivos generales sobre la media
df['mean'].describe()
```

```python
# Histograma de la Tmin promedio
plt.figure(figsize=(10,6))
sns.histplot(df['mean'], kde=True, bins=30)
plt.title('Distribución de la temperatura mínima promedio por distrito')
plt.xlabel('Temperatura mínima promedio (°C)')
plt.ylabel('Frecuencia')
plt.show()
```
